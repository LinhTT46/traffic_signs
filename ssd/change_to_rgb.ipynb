```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
import glob
```

```python

def change_to_rgb(pathfile):
    for file in glob.glob(pathfile):
        image = cv2.imread(file)
        image = cv2.resize(image, (320, 240)) 
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        cv2.imwrite(file, image)
        
```

```python
change_to_rgb('/home/li/Traffic_Signs/datasets/Images/train/*.png')
```

```python
change_to_rgb('/home/li/Traffic_Signs/datasets/Images/val/*.png')
```
